## **Question: Correlation between price and number of reviews: Is there a relationship between the price of a product and the number of reviews it has?**

### LIBRARIES

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset path
DATA_PATH = './Data/Data.csv'

# Load the dataset
data = pd.read_csv(DATA_PATH)
```

```python
# Filter the data to remove rows with missing or zero values in relevant columns
filtered_data = data[(data['Giá'] > 0) & (data['Lượt đánh giá'] > 0)]

# Create a scatter plot to visualize the correlation between price and number of reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(x=filtered_data['Giá'], y=filtered_data['Lượt đánh giá'], alpha=0.7)
plt.title("Correlation Between Price and Number of Reviews", fontsize=14)
plt.xlabel("Price (VND)", fontsize=12)
plt.ylabel("Number of Reviews", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Calculate correlation matrix for numerical columns
numerical_columns = ['Giá', 'Lượt đánh giá', 'Điểm đánh giá trung bình', 'Số lượng đã bán']  # Replace with English column names if available
correlation_matrix = data[numerical_columns].corr()

# Create a heatmap to visualize correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Heatmap of Correlations Between Numerical Features", fontsize=14)
plt.show()
```

### **Analysis and Visualization Remarks:**

#### Scatter Plot (Correlation between Price and Number of Reviews):
- The scatter plot shows that the relationship between product price (**Giá**) and the number of reviews (**Lượt đánh giá**) is not clear.
- The number of reviews does not significantly increase with higher product prices. Most products with low to medium prices (below 1 million VND) tend to receive more reviews.
- Products with higher prices do not necessarily receive more reviews, which might be because higher prices do not attract as many customers.

#### Heatmap (Correlation Matrix of Variables):
- The correlation between **Price** and **Number of Reviews** is quite low (~0.12), indicating that price does not significantly affect the number of reviews.
- **Average Rating** (**Điểm đánh giá trung bình**) does not show strong relationships with other variables either.
- A negative correlation (-0.17) between **Price** and **Units Sold** suggests that higher-priced products tend to sell less compared to lower-priced ones.

### **Conclusion:**
- The number of reviews does not significantly influence the price of a product.
- Other factors (such as brand reputation, product category, or marketing strategies) are likely to play a more significant role in determining the price.
